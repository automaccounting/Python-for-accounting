```python
import pandas as pd
wac_df = pd.read_excel(r'C:\Users\Davide\Desktop\Python for Accounting workspace\cmp.xlsx', skiprows=4)
```

```python
pvt = pd.pivot_table(wac_df,
               index= ['REF.', 'DESC.REF'], 
               values=['QUAN','SUM TOT'],
               aggfunc='sum')

pvt['WAC'] = pvt['SUM TOT']/pvt['QUAN']
```
